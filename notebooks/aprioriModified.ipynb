```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
import time
import psutil



# Define the minimum support levels
min_support_levels = [500, 1000, 1500, 2000, 2500] #[100, 150, 200, 250, 300]

# Initialize list to store results dictionaries
results = []

# Iterate over minimum support levels
for min_sup in min_support_levels:
    start_time = time.time()
    
    # Read the dataset
    df = pd.read_csv('Transactional_T10I4D100K (2).csv', header=None)
    
    # Preprocess the dataset
    preprocessed_data = []
    for transaction in df[0]:
        items = transaction.split('\t')
        preprocessed_data.append(items)
    
    # Encode the dataset
    te = TransactionEncoder()
    te_ary = te.fit_transform(preprocessed_data)
    df_encoded = pd.DataFrame(te_ary, columns=te.columns_)

    # Extract frequent patterns using Apriori algorithm with low_memory=True
    frequent_itemsets = apriori(df_encoded, min_support=min_sup/len(df_encoded), use_colnames=True, low_memory=True)
    
    end_time = time.time()
    runtime = end_time - start_time
    
    # Get memory usage
    memory = psutil.Process().memory_info().rss
    
    # Append results dictionary to list
    results.append({
        'algorithm': 'Apriori',
        'minSup': min_sup,
        'patterns': len(frequent_itemsets),
        'runtime': runtime,
        'memory': memory
    })

# Convert results list into DataFrame
results_df = pd.DataFrame(results)

# Print the result
print(results_df)
```

```python
from PAMI.frequentPattern.basic import Apriori as alg
import pandas as pd

# Set the input parameters
inputFile = 'Transactional_T10I4D100K (2).csv'
separator = '\t'
minimumSupportCountList = [500, 1000, 1500, 2000, 2500]

# Initialize a data frame to store the results of Apriori
result = pd.DataFrame(columns=['algorithm', 'minSup', 'patterns', 'runtime', 'memory'])

# Execute the algorithm at different minSup values
for minSupCount in minimumSupportCountList:
    obj = alg.Apriori(inputFile, minSup=minSupCount, sep=separator)
    obj.startMine()
    # Store the results in the data frame
    result.loc[result.shape[0]] = ['Apriori', minSupCount, len(obj.getPatterns()), obj.getRuntime(), obj.getMemoryRSS()]

# Print the results
print(f"Frequent patterns were generated successfully using Apriori algorithm")

print(result)
```
